```python
# David Bell

from sympy import *
from sympy.plotting import (plot, plot_parametric,plot3d_parametric_surface, plot3d_parametric_line,plot3d)
from sympy import pi
# Part 1
t=symbols('t',positive=True)
x1=t**2
y1=t**3-3*t
x1y1 = x1-y1

#Part a
pltcurves=plot_parametric((x1,y1,(t,-2,2))) 

#Part B

x=diff(x1,t)
y=diff(y1,t)
tanvec1 = Matrix([x.subs(t,-sqrt(3)),y.subs(t,-sqrt(3))])
tanvec2 = Matrix([x.subs(t,sqrt(3)),y.subs(t,sqrt(3))])

u1=tanvec1/tanvec1.norm()
u2=tanvec2/tanvec2.norm()

print('the tan vect is:',tanvec1,'and',tanvec2)

#Part C
theta = acos(tanvec1.dot(tanvec2)/(tanvec2.norm()*tanvec2.norm()))
theta *= (180/pi)
print("the angle is ",theta)
```

```python
# Part 2

t=symbols('t')

x= 3*t**2+1
y= 2*t**3+1

#Part a
G=plot_parametric(x,y,(t,-2.5,2.5), xlim = [0,15],ylim = [-15,10],show =False)


#Part B
dxdt= diff(x,t)
dydt= diff(y,t)

dxdy=dydt/dxdt
line = dxdy*(x-4)+ 3-y
ans = solve(line,t)

print(ans)

x,y=symbols('x,y')
b = ans[0]*(x-4)+3
a = ans[1]*(x-4)+3

G1 =plot(a,(x,0,15),ylim=[-15,10], show = False)
G2 =plot(b,(x,0,15),ylim=[-15,10], show = False)
G.extend(G1)
G.extend(G2)
G.show()

x,y=symbols('x,y')
print(ans[0]*(x-4)+3)
print(ans[1]*(x-4)+3)
```

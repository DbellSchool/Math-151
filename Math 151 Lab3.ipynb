```python
# Part 1
x = sp.Symbol('x')

              
# defines function
g = ((x**2)*sp.sin(1/x))
f = (-x**2)
h = (x**2)

# A

# finds limit
L = sp.limit(g,x,0) 
print("The limit as x-->0 is: ",L)

# B

# graphs
graphG = spp.plot_parametric(x,g,(x,-.5,.5),Ylim = [-5,5],show = False)
graphF = spp.plot_parametric(x,f,(x,-.5,.5),Ylim = [-5,5],show = False)
graphH = spp.plot_parametric(x,h,(x,-.5,.5),Ylim = [-5,5],show = False)
# adds graphs together
graphG.extend(graphF)
graphG.extend(graphH)
# shows 
graphG.show()


# C
print("\nThis ilistates the Squeeze therom")

```

```python
# Lab 2
import sympy as sp
import sympy.plotting as spp


#Part 1
t = sp.Symbol('t')
f = (1-sp.exp(-t))/(1+sp.exp(-t))
graph = spp.plot_parametric(t,f,(t,-5,5),Ylim = [-5,5])
# part a
print("This fuction passes the horasontal line test")

# part b
f2 = -sp.ln((1+t)/(-t+1))
graph2 = spp.plot_parametric(t,f2,(t,-5,5),Ylim = [-5,5])

# part C
print("it is the inverse")
# part d
```

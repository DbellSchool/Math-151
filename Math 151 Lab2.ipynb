```python
# Part 2


x = sp.Symbol('x')
f = (sp.log(x)-sp.log(7))/(x-7)

# part a
print ("\nPart A \n  leftval\n\n")
leftval = [print(f.subs({x:i}).evalf()) for i in (6.9,6.99,6.99)]  
print("\n  rightval\n\n")
rightval = [print(f.subs({x:i}).evalf()) for i in (7.1,7.01,7.001)]

# part b
print("\nPart B")
sp.plot(f,(x,6,8))

# part c
print("\nPart C")
print(f.subs({x:7.00000000001}).evalf(),end="")
print(" =",f.subs({x:6.99999999999}).evalf())
print("and at 7 returns -",f.subs({x:7}).evalf(),"-")

```

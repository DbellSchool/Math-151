```python
# Part 1
import sympy as sp
import sympy.plotting as spp

def reciprocal(n):# allows for eisly taking recirical
    return 1.0 / n

#A
x = sp.Symbol('x')
y = sp.Symbol('y')
f = (5*x)/ (1+x**2)

yVal =f.subs(x, 2)

fdy= sp.diff(f,x)
print ("The dirivative is",fdy)

m1 =  fdy.subs(x, 2)
print ("The slop at x= 2 is",m1)


# A
ftan = m1*(x-2) + yVal

# B
opRecip = -1*reciprocal(m1)
print(m1,opRecip)

#C
fperp = opRecip*(x-2) + yVal
g = sp.plot((f,(x,0,4)),(fperp,(x,1,3)),(ftan,(x,1,3)))

```

```python
#Part 2

t = sp.Symbol('t')
s = 1+10*t-1.86*t**2
v = sp.diff(s,t)
# A
V  = sp.solve(v,t) # solves for v(t)
yV = s.subs(t, V[0])

# A 
Vel_oneSec = v.subs(t, 1)
print("\nThe Velosity after one second is", Vel_oneSec)
# part B
print("\nAt time ",V[0], "s The ball is at the highest Point of",yV,'ft')

# part C
posZero  = sp.solve(s,t) # solves for v(t)
Vel_Ground = v.subs(t,posZero[1] )
print("\nThe rock will hit the ground at t =",posZero[1], "with a velocity of", Vel_Ground)
```

```python
#Part 3

# A/B
a = sp.Symbol('a')
F = x**2
G = a*x**2-2*x+2

FPrime = sp.diff(F,x)
GPrime = sp.diff(G,x)

ans = sp.solve([FPrime-GPrime,F-G],x,a)

print(ans) ## a = 3/2 # x = 2


#C

F = x**2
G = (3/2)*x**2-2*x+2 # pluged in a val from part a

FPrime = sp.diff(F,x)
GPrime = sp.diff(G,x)

# slope equals x @ 2
tan_slope = GPrime.subs(x, 2)
yval = F.subs(x,2)

print(tan_slope)
print(FPrime,GPrime)


tanLine = tan_slope*(x-(2))+yval

sp.plot((F,(x,1,3)),(G,(x,1,3)),(tanLine,(x,1,3)))
```
